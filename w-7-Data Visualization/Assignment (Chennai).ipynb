```python
zomato_df = raw_df.copy()
```

```python
zomato_df.head()
```

```python
fig = px.histogram(zomato_df, 
                   x='Cuisine', 
                   color='Cuisine', 
                   title='No. of Restaurants by Cuisine Type', 
                   labels={'Cuisine':'Cuisine'})
fig.show()
```

```python
rating_type_df = zomato_df['Dining Rating'].value_counts().reset_index()
rating_type_df.rename(columns={'index':'Dining Rating', 'Dining Rating':'COUNT OF RESTAURANTS'}, inplace=True)
rating_type_df
```

```python
rating_type_df.columns = ['COUNT OF RESTAURANTS','Dining Rating']
```

```python
fig = px.pie(
    rating_type_df,
    names='Dining Rating',
    values='COUNT OF RESTAURANTS',
    title='Percentage of Restaurants by Dining Rating'
)
fig.show()
```

```python
highest_rated_df = zomato_df[zomato_df['Dining Rating'] >= 4.5]
highest_rated_df
```

```python
fig = px.histogram(highest_rated_df, x='Cuisine', color='Location',
 title= 'No. of Best Restaurant for each Cuisine Type by Places')
# .update_xaxes(categoryorder="total desce")
fig.show()
```

```python
highest_rated_price_df = highest_rated_df.groupby(by=['Location', 'Cuisine'])['Price for 2'].mean().reset_index()
highest_rated_price_df.head()
```

```python
chinese_df = zomato_df[zomato_df['Cuisine'].str.contains('Chinese')]
chinese_df
```

```python
fig = px.bar(
    chinese_df, 
    x='Location', 
    y='Dining Rating Count', 
    color='Dining Rating', 
    title= 'No. of Chinese Restaurant by Location')
fig.show()
```
