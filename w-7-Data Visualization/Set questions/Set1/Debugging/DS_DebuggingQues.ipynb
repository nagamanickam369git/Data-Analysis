```python
import numpy as np
from scipy.stats import zscore

data = [1, 2, 3, 100, 5, 6, 7, 8, 9]
z_scores = zscore(data)
# print(z_scores)
outliers=[]
for i in range(len(data)):
    # print(abs(z_scores[i]))
    if abs(z_scores[i]) > 2:
        outliers.append(i)


print('outliers boundries',outliers)
print('outliers',data[outliers[0]])
```

```python
import numpy as np
from scipy.stats import zscore

data = np.array([1, 2, 3, 100, 5, 6, 7, 8, 9])
median = np.median(data)
print(median)
z_scores = zscore(data)
print(z_scores)

for i in range(len(data)):
    if abs(z_scores[i]) > 2:
        data[i] = median

print("Cleaned data:", data)
# data = [median if abs(zscore(x)) > 2 else x for x in range(len(data))]
```
