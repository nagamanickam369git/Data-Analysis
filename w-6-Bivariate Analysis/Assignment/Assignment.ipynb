### 7)Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)

```python
from scipy.stats import f_oneway

# Significance level
alpha = 0.05

group1 = dataset['etest_p']
group2 = dataset['mba_p']

# Perform ANOVA
f_statistic, p_value = f_oneway(group1, group2)

print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")

# Decision
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (H0)")
    print("Conclusion: There is a statistically significant difference between etest_p and mba_p at 5% significance level.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
    print("Conclusion: There is no statistically significant difference between etest_p and mba_p at 5% significance level.")
```

### 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)

```python
from scipy.stats import ttest_ind

# Significance level
alpha = 0.05

# Select salary groups
group1t = dataset[dataset['degree_t'] == 'Sci&Tech']['salary']
group2t = dataset[dataset['specialisation'] == 'Mkt&HR']['salary']

# Perform Independent t-test
t_statistic, p_value = ttest_ind(group1t, group2t, nan_policy='omit')

# Print test values
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")

# Decision
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (H0)")
    print(
        "Conclusion: There is a statistically significant difference in salary "
        "between degree_t (Sci & Tech) and specialisation (Mkt & HR) "
        "at 5% significance level."
    )
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
    print(
        "Conclusion: There is no statistically significant difference in salary "
        "between degree_t (Sci & Tech) and specialisation (Mkt & HR) "
        "at 5% significance level."
    )
```

### 9)Convert the normal distribution to standard normal distribution for salary column

```python
# Calculate mean and std
salary_mean = dataset['salary'].mean()
salary_std = dataset['salary'].std()

# z=x−μ/σ
# x = salary value
# μ = mean of salary
# σ = standard deviation of salary

# Convert to standard normal (Z-score)
dataset['salary_z'] = (dataset['salary'] - salary_mean) / salary_std

print(dataset['salary_z'].mean())
print(dataset['salary_z'].std())
```

```python
from scipy.stats import zscore

dataset['salary_z'] = zscore(dataset['salary'])
dataset['salary_z']
```

### 10)What is the probability Density Function of the salary range from 700000 to 900000?

```python
from scipy.stats import norm

mu = dataset['salary'].mean()
sigma = dataset['salary'].std()

# Probability between 700000 and 900000
probability = norm.cdf(900000, mu, sigma) - norm.cdf(700000, mu, sigma)

print("Probability:", probability)
```

#### Probability close to 0 so very rare


### 11)Test the similarity between the degree_t(Sci&Tech) with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hy-pothesis Testing)

```python
group = dataset[dataset['degree_t'] == 'Sci&Tech']

etest = group['etest_p']
mba = group['mba_p']
```

```python
from scipy.stats import ttest_rel
# Significance level
alpha = 0.05

t_stat, p_value = ttest_rel(etest, mba)

# Print test values
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Decision
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (H0)")
  
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
  
```
