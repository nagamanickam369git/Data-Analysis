```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
dataset
```

```python
from Univariate import UnivariateBuilder
quan,qual = UnivariateBuilder.quanQualFinder(dataset)
print("quan:",quan)
print("qual:",qual)
```

```python
descriptive = pd.DataFrame()
descriptive
```

```python
descriptive = pd.DataFrame(index = ["Mean","Median","Mode"],columns=quan)
descriptive
```

```python
for colName in quan:
    descriptive[colName]['Mean'] = dataset[colName].mean()
    descriptive[colName]['Median'] = dataset[colName].median()
    descriptive[colName]['Mode'] = dataset[colName].mode()[0]
    
```

### Understood from POC

##### Mean, Median, Mode values we can find only for interger or float datatype only.

#### Find Informations about it.

#### Mean value:
##### Mean to find average.
The candidate has scored 67% in SSC, 66% in HSC, 66% in their degree, 72% in the entrance test (improved), and 62% in MBA (slightly decreased / average performance).

Note: Don't consider salary related values in mean method becasue it calculated with outliers.

#### Median value:
##### Median NOT allow outliers.
The median salary is 265000.0.

#### Mode value:
##### Mode is used to find repeated values.
The candidate repeatedly scored similar marks throughout their academics: 62 marks in SSC, 63 marks in HSC, 65 marks in their degree, 60 marks in the entrance test, and 56 marks in MBA.
