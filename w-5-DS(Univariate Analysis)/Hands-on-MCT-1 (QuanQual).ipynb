```python
quan=[]
qual=[]
for colName in dataset.columns:
    if dataset[colName].dtype =='O':
        qual.append(colName)
    else:
        quan.append(colName)


print("quan cols:",quan)
print("qual cols:",qual)
```

```python
def quanQualFinder(dataset):
    quan=[]
    qual=[]
    for colName in dataset.columns:
        if dataset[colName].dtype =='O':
            qual.append(colName)
        else:
            quan.append(colName)
    return quan,qual
    
    
```

```python
quanQualFinder(dataset)
```
